# Import library and model

```python
import cv2 as cv
from matplotlib import pyplot as plt
```

```python
face_cascade = cv.CascadeClassifier('haarcascade_frontalface_default.xml')
```

# Detect face in image

```python
def detect(gray,frame):
    faces = face_cascade.detectMultiScale(gray,1.3,5)
    for (x,y,w,h) in faces:
        cv.rectangle(frame, (x, y), (x+w, y+h), (255, 0, 0), 2)
        
    return frame
    
#
ima_cv = cv.imread('girl_face.jpeg',1)
ima = cv.cvtColor(ima_cv, cv.COLOR_BGR2RGB)
#
print("Input: ")
plt.imshow(ima)
plt.show()
#
gray = cv.cvtColor(ima, cv.COLOR_BGR2GRAY)
result = cv.cvtColor(detect(gray,ima),cv.COLOR_BGR2RGB)
# 
print("Result: ")
plt.imshow(result)
plt.show()

```
